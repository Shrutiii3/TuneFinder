```python
import numpy as np 
import pandas as pd
import langid
```

```python
df = pd.read_csv('songdata.csv')
df.head()
```

```python
df.shape
```

```python
df = df[['artists','album_name','track_name','track_genre']]
df
```

```python
def is_english(text):
    if not isinstance(text, str) or text.strip() == '':
        return False
    lang, _ = langid.classify(text)
    return lang == 'en'


# Filter the DataFrame
```

```python
# Filter the DataFrame
df = df[df['track_name'].apply(is_english)]
```

```python
df.head()
```

```python
df = df.sample(n=5000).reset_index(drop=True)
df
```

```python
df['tag'] = df['track_name'] +" "+ df['artists'] 
df.head()
```

```python
df['tag'] = df['tag'] +" "+ df['album_name']
df.head()
```

```python
df['tag'] = df['tag'] +" "+ df['track_genre']
df.head()
```

```python
df['tag'] = df['tag'].str.lower().replace(r'[^\w\s]','').replace(r'\n',' ',regex=True)
df['tag'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
y=[]

def tokenization(text):
    tokens = nltk.word_tokenize(text)
    y=[ps.stem(word) for word in tokens]
    
    return " ".join(y)
    
```

```python
df['tag'] = df['tag'].apply(lambda x : tokenization(x))
```

```python
df['tag'][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfid = TfidfVectorizer(stop_words = 'english')
matrix = tfid.fit_transform(df['tag'])
```

```python
matrix.shape
```

```python
similarity_mat = cosine_similarity(matrix)
```

```python
similarity_mat
```

```python
def recommend(song):
    index = df[df['track_name']==song].index[0]
    distances = similarity_mat[index]
    song_list = sorted(list(enumerate(distances)),reverse=True,key= lambda x :x[1])[1:11]
    
    for i in song_list:
        print(df.iloc[i[0]].track_name)
```

```python
recommend('Left Behind')
```
